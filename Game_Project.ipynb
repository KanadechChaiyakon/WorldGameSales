```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
game = pd.read_csv('vgsales.csv',encoding='latin1')
```

```python
game
```

## นำเกมที่ไม่แสดงค่า Year ออกไม่นำมาคิด

```python
game = game.drop(game.loc[game['Year'].isna()].index)
```

## นำผลการสำรวจในปี 2020 และ 2017 ออกเนื่องจากสำรวจข้อมูลน้อย(1 และ 3 แถว ตามลำดับ)

```python
game = game.drop(game.loc[(game['Year'] == 2017) | (game['Year'] == 2020)].index)
```

```python
game
```

```python
print(game['Platform'].unique() , '\n\nTotal Platform : ' , len(game['Platform'].unique()))
```

```python
print(game['Genre'].unique() , '\n\nTotal Genre : ' , len(game['Genre'].unique()))
```

```python
print(game['Publisher'].unique() , '\n\nTotal Publisher : ' , len(game['Publisher'].unique()))
```

# 1

```python
game_na = game.set_index(['Year','Genre']).sort_values('NA_Sales',ascending=False).sort_index(level=0,sort_remaining=False)
```

```python
game_eu = game.set_index(['Year','Genre']).sort_values('EU_Sales',ascending=False).sort_index(level=0,sort_remaining=False)
```

```python
game_jp = game.set_index(['Year','Genre']).sort_values('JP_Sales',ascending=False).sort_index(level=0,sort_remaining=False)
```

```python
game_na
```

```python
game_na_top = game_na.groupby(level=0).head(1)
```

## เกมที่มียอดขายดีที่สุดใน NA ในแต่ละปี

```python
game_na_top
```

```python
game_eu_top = game_eu.groupby(level=0).head(1)
```

```python
game
```

```python
game_2_gta = game.loc[game['Name'].str.contains('Grand Theft Auto')]
```

```python
game_2_gta_outer = game.loc[game['Name'].str.contains('Grand Theft Auto'),['Name','Year']].drop_duplicates(subset="Name")
```

```python
gta_result = game_2_gta.groupby('Name').sum().drop('Year',axis=1).join(game_2_gta_outer.set_index('Name')).sort_values('Year').loc[:,['Year','Global_Sales']].reset_index()
```

```python
gta_result
```

```python
game_2_final = game.loc[game['Name'].str.contains('Final Fantasy')]
```

```python
game_2_final_outer = game.loc[game['Name'].str.contains('Final Fantasy'),['Name','Year']].drop_duplicates(subset="Name")
```

```python
final_result = game_2_final.groupby('Name').sum().drop('Year',axis=1).join(game_2_final_outer.set_index('Name')).sort_values('Year').loc[:,['Year','Global_Sales']].reset_index()
```

```python
final_result
```

```python
game_2_ac = game.loc[(game['Name'].str.contains('Assassin\'s Creed')) & (game['Publisher'] == 'Ubisoft')].sort_values('Year')
```

```python
game_2_ac_outer = game.loc[game['Name'].str.contains('Assassin\'s Creed'),['Name','Year']].drop_duplicates(subset="Name")
```

```python
ac_result = game_2_ac.groupby('Name').sum().drop('Year',axis=1).join(game_2_ac_outer.set_index('Name')).sort_values('Year').loc[:,['Year','Global_Sales']].reset_index()
```

```python
ac_result
```

```python
gta_result.plot.barh(x='Name', y='Global_Sales')
final_result.plot.barh(x='Name', y='Global_Sales',figsize=(10, 14))
ac_result.plot.barh(x='Name', y='Global_Sales')
```
